```python
import pandas as pd
import numpy as np
import panel as pn
import matplotlib as plt
#creates interactive tables
pn.extension('tabulator')

import hvplot.pandas
```

```python
#reading the dataset into df
df = pd.read_csv('billionaires_statistics_dataset.csv')
```

```python
#filling all the na values with 0
df = df.fillna(0)
len(df)
#making df interactive
idf = df.interactive()
```

```python
#making a select widget and putting all the countries in it
options1 = []
for i in range(len(df)):
    if df.country[i] in options1:
        continue
    else:
        options1.append(df.country[i])
optoins1 = options1.remove(0)
options1 = sorted(options1)
    
yaxis_countries = pn.widgets.Select(
    name='Countries', options=['France']
    
)
yaxis_countries.options = options1
yaxis_countries.options
pn.Row(yaxis_countries).show()
```

```python
#seeing all the categories we have 
categories = []
for i in range(len(df)):
    if df.category[i] in categories:
        continue
    else:
        categories.append(df.category[i])
categories
```

```python
df['finalWorth'] = df['finalWorth'].astype(int)
```

```python

graph1_val = df.groupby('category')['finalWorth'].sum().reset_index()
graph1_val
graph1_val_chart = graph1_val.hvplot.bar(
    x = 'category',
    y = 'finalWorth',
    height = 250,
    width = 2500,
    title = 'Category and Wealth',
)
graph1_val_chart
    
```

```python
#add up the finalWorth for each category
dict = {}
for i in range(len(df)):
    if df.category[i] in dict:
        dict[df.category[i]] = dict[df.category[i]] + df.finalWorth
    else:
        dict[df.category[i]] = df.finalWorth
dict

```
